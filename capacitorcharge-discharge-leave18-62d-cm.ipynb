# Capacitor charging / discharging

### Introduction 

As a capacitor of capacitance C is discharged through a resistor of resistance R the voltage across the capacitor V varies according to
$V = V_{0}e^{−t/RC}$ (1) where V0 is the voltage across the capacitor when t = 0.
If the same capacitor is now charged in series with a resistance R, the voltage across the capacitor varies according to
$V =V{(1−e^{−t/RC})}$  (2) 0
where V0 is the voltage across the capacitor when it is fully charged (i.e. the supply voltage).

$V = V_{0} e^{-t/RC}$ 


### Method 

 

The circuit diagaram as shown was made. The capacitor was charged until the reading on the voltmeter did not increase. The capacitor was discharged. The voltage was measured for one minute at 10 second intervals. Once the voltage through the capacitor was circa 0, the capacitor was charged. The voltage was measured for one minute at 10 second intervals.
